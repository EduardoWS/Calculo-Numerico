**(c) Encontre o coeficiente de variação amostral em cada grupo.**

```python
# CÓDIGO PYTHON:

import numpy as np


# Dados dos fumantes
fumantes = [69.3, 56.0, 22.1, 47.6, 53.2, 48.1, 52.7, 34.4, 60.2, 43.8, 23.2, 13.8]
media_fumantes = np.mean(fumantes)  # Calcula a média dos fumantes
desvio_padrao_fumantes = np.std(fumantes)  # Calcula o desvio-padrão dos fumantes
coef_var_fumantes = (desvio_padrao_fumantes / media_fumantes) * 100  # Calcula o coeficiente de variação dos fumantes



# Dados dos não-fumantes
nao_fumantes = [28.6, 25.1, 26.4, 34.9, 29.8, 28.4, 38.5, 30.2, 30.6, 31.8, 41.6, 21.1, 36.0, 37.9, 13.9]
media_nao_fumantes = np.mean(nao_fumantes)  # Calcula a média dos não-fumantes
desvio_padrao_nao_fumantes = np.std(nao_fumantes)  # Calcula o desvio-padrão dos não-fumantes
coef_var_nao_fumantes = (desvio_padrao_nao_fumantes / media_nao_fumantes) * 100  # Calcula o coeficiente de variação dos não-fumantes



# Imprime os resultados
print(f"Coeficiente de variação amostral dos fumantes: {coef_var_fumantes:.2f}%")
print(f"Coeficiente de variação amostral dos não-fumantes: {coef_var_nao_fumantes:.2f}%")
```

**7) A espessura (mm) de 9 parafusos foi mensurada e os dados obtidos foram:**


**2.8; 4.2; 5.5; 3.0; 4.1; 3.9; 2.7; 4.2 e 2.9.**


**Calcule  as  medidas  descritivas: mínimo, máximo, quartis, mediana, média, amplitude, desvio  médio, variância, desvio padrão e coeficiente de variação.**


```python
# CÓDIGO PYTHON:

import numpy as np
import statistics



# Dados dos parafusos
dados = [2.8, 4.2, 5.5, 3.0, 4.1, 3.9, 2.7, 4.2, 2.9]

# Colocando em ordem crescente
dados_ordenados = sorted(dados)
print("Dados ordenados:", dados_ordenados)
print()

# Medidas descritivas
minimo = np.min(dados_ordenados)
maximo = np.max(dados_ordenados)
quartis = np.percentile(dados_ordenados, [25, 50, 75])
mediana = np.median(dados_ordenados)
media = np.mean(dados_ordenados)
amplitude = maximo - minimo
desvio_medio = statistics.mean([abs(x - media) for x in dados_ordenados])
variancia = np.var(dados_ordenados)
desvio_padrao = np.std(dados_ordenados)
coef_variacao = (desvio_padrao / media) * 100

# Imprimir os resultados
print("Mínimo:", minimo)
print("Máximo:", maximo)
print("Quartis:", quartis)
print("Mediana:", mediana)
print("Média:", media)
print("Amplitude:", amplitude)
print("Desvio Médio:", desvio_medio)
print("Variância:", variancia)
print("Desvio Padrão:", desvio_padrao)
print("Coeficiente de Variação:", coef_variacao)
```

**(b) Estabeleça a distribuição de frequência relativas, acumulada e acumulada relativa.**

```python
# CÓDIGO PYTHON:

import pandas as pd

# Dados de vida útil das bombas de combustível
dados = [2.0, 3.0, 0.3, 3.3, 1.3, 0.4, 0.2, 6.0, 5.5, 6.5,
         0.2, 2.3, 1.0, 6.0, 5.6, 1.5, 4.0, 5.9, 1.8, 4.7,
         0.7, 4.5, 0.3, 1.5, 0.5, 2.5, 5.0, 6.0, 1.2, 0.2]

# Criar um DataFrame para os dados
df = pd.DataFrame(dados, columns=['Vida útil'])

# Calcular a distribuição de frequência relativa
frequencia_relativa = df['Vida útil'].value_counts(normalize=True).sort_index()

# Calcular a distribuição de frequência acumulada
frequencia_acumulada = frequencia_relativa.cumsum()

# Calcular a distribuição de frequência acumulada relativa
frequencia_acumulada_relativa = frequencia_acumulada / frequencia_acumulada.max()

# Imprimir as distribuições de frequência
print("Distribuição de Frequência Relativa:")
print(frequencia_relativa)
print("\nDistribuição de Frequência Acumulada:")
print(frequencia_acumulada)
print("\nDistribuição de Frequência Acumulada Relativa:")
print(frequencia_acumulada_relativa)
```

A saída do código mostra as distribuições de frequência relativa, acumulada e acumulada relativa para os dados de vida útil das bombas de combustível.

A distribuição de frequência relativa mostra a proporção de cada valor em relação ao total de observações. A distribuição de frequência acumulada indica a soma cumulativa das frequências relativas. A distribuição de frequência acumulada relativa mostra a proporção acumulada de cada valor em relação ao total de observações.

Essas distribuições fornecem uma visão mais detalhada da distribuição dos dados de vida útil das bombas de combustível, permitindo analisar a frequência relativa de cada valor e a acumulação dessas frequências ao longo do conjunto de dados.

**(c) Calcule a média, a amplitude e o desvio padrão amostrais.**

```python
# CÓDIGO PYTHON

import numpy as np

# Dados de vida útil das bombas de combustível
dados = [2.0, 3.0, 0.3, 3.3, 1.3, 0.4, 0.2, 6.0, 5.5, 6.5,
         0.2, 2.3, 1.0, 6.0, 5.6, 1.5, 4.0, 5.9, 1.8, 4.7,
         0.7, 4.5, 0.3, 1.5, 0.5, 2.5, 5.0, 6.0, 1.2, 0.2]

# Média amostral
media_amostral = np.mean(dados)

# Amplitude amostral
amplitude_amostral = np.max(dados) - np.min(dados)

# Desvio padrão amostral
desvio_padrao_amostral = np.std(dados, ddof=1)

# Imprimir os resultados
print("Média Amostral:", media_amostral)
print("Amplitude Amostral:", amplitude_amostral)
print("Desvio Padrão Amostral:", desvio_padrao_amostral)
```
